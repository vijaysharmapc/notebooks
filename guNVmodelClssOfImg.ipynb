```python
from sklearn.datasets import load_digits
import seaborn as sns


digits = load_digits()
digits.images.shape # 1797 images ,each consisiting of an 8*8 grid of pixels
```

```python
type(digits)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig,axes = plt.subplots(10,10,figsize=(8,8),subplot_kw={'xticks':[],'yticks':[]}
                        ,gridspec_kw=dict(hspace=0.1,wspace=0.1))
for i,ax in enumerate(axes.flat):
    ax.imshow(digits.images[i],cmap='binary',interpolation='nearest')
    ax.text(0.05,0.05,str(digits.target[i]),
           transform=ax.transAxes,color='green')
```

```python
X = digits.data # flattened out pixels-array of length 64
X.shape
```

```python
y =  digits.target
y.shape
```

```python
# 1,797 samples and 64 features
```

```python
# Classificaton of digits
```

```python
from sklearn.cross_validation import train_test_split
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(Xtrain,ytrain)
y_model = model.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,y_model)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ytest,y_model)
sns.heatmap(mat,square=True,annot=True,cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value')
# large numbe of 2's are miss classified as 1 or 8
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
x_iris =  iris.drop('species',axis=1)
```

```python
x_iris.head()
```

```python
x_iris.shape # 150 rowsm4 columns
```

```python
y_iris = iris['species']
y_iris.shape
```

```python
y_iris[0:5]
```

```python
from sklearn.cross_validation import  train_test_split
Xtrain,Xtest,ytrain,ytest = train_test_split(x_iris,y_iris,random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
model.fit(Xtrain,ytrain)
y_model = model.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,y_model)
```

```python
# use other visual things
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ytest,y_model)
sns.heatmap(mat,square=True,annot=True,cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value')
```
